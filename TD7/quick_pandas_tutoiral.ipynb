```python
def consider_ot_payment(salary):
  result = ""
  if salary>70000:
    result = "NO"
  elif salary>50000:
    result = "MAY BE"
  else:
    result = "YES"

  return result
```

```python
consider_ot_payment(80000)
```

```python
consider_ot_payment(60000)
```

```python
consider_ot_payment(30000)
```

```python
df["salary"].apply(consider_ot_payment)
```

```python
df["should_pay_ot"] = df["salary"].apply(consider_ot_payment)
```
