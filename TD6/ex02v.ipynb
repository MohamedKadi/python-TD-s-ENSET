```python
import numpy as np
import matplotlib.pyplot as plt

def simul_X():
    return np.random.binomial(n=3, p=1/3)

n_simulations = 10000
results = [simul_X() for _ in range(n_simulations)]

plt.figure(figsize=(10, 6))
plt.hist(results, bins=[-0.5, 0.5, 1.5, 2.5, 3.5], edgecolor='black')
plt.xlabel('nombre des pile')
plt.ylabel('frequency')
plt.title('graph')
plt.xticks([0, 1, 2, 3])
plt.show()
```

```python
import numpy as np
import numpy.random as rd

def T(n):
    S = 0
    T = 0
    while S < n:
        tirage = rd.randint(1, n+1)
        S = S + tirage
        T = T + 1
    return T
```
